```python
import matplotlib.pyplot as plt

# Data from the first run (Data Set 1)
n_values = [10, 100, 1000, 10000, 100000]
experimental = [66995.62, 22888.18, 24795.53, 125885.01, 1352310.18]
adjusted_theoretical = [92317.11, 338079.90, 641759.82, 983051.19, 1352310.18]

# Plot
plt.figure(figsize=(10, 6))
plt.plot(n_values, experimental, 'r-', label='Experimental Results', marker='o')
plt.plot(n_values, adjusted_theoretical, 'b-', label='Adjusted Theoretical Results', marker='x')

# Use logarithmic scales for better comparison
plt.xscale('log')
plt.yscale('log')

# Labels and title
plt.xlabel('n (log scale)')
plt.ylabel('Time (ns, log scale)')
plt.title('Comparison of Experimental and Adjusted Theoretical Results')
plt.legend()

# Display grid and the plot
plt.grid(True)
plt.show()
```
